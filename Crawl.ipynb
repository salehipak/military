```python
from requests import get
from warnings import warn
from time import sleep
from random import randint
import numpy as np
import pandas as pd
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')         # to fliter warning in notebook
```

## Data Collecting

```python
# Demand
dmd_id = []
dmd_title = []
dmd_urlIdentifier = []
dmd_description = []
dmd_key_words = []
dmd_expiration_date = []

for i in np.arange(1,939): #939
    try:
        if i%10 == 0:
            sleep(1)
            print(i)
        url = "https://techmart.ir/api/demand/public/DMD-" + str(i) #+"?vtPreventCache=1693470146016"
        response = get(url,verify=False)
        js = response.json()
        dmd_id.append(js['uid'])
        dmd_title.append(js['title'])
        dmd_urlIdentifier.append(js['urlIdentifier'])
        dmd_description.append(js['description'])
        dmd_key_words.append( [js['key_words'][_]['title'] for _ in range(len(js['key_words']))])
        dmd_expiration_date.append(datetime.fromtimestamp(js['expiration_date']/1000))
    except:
        continue
```

```python
# Supply
prd_id = []
prd_title = []
prd_urlIdentifier = []
prd_description = []
prd_key_words = []
prd_owner_id =[]
prd_owner_type=[]
prd_owner_name=[]

for i in np.arange(1,17435): #17435
    if i%10 == 0:
        sleep(1)
        print(i)
    try:
        url = "https://techmart.ir/api/product/public/PRD-"+str(i) #+"?vtPreventCache=1693472804284"
        response = get(url,verify = False)
        js = response.json()
        prd_id.append(js['uid'])
        prd_title.append(js['title'])
        prd_urlIdentifier.append(js['urlIdentifier'])
        prd_description.append(js['description'])
        prd_key_words.append( [js['key_words'][_]['title'] for _ in range(len(js['key_words']))])
        prd_owner_id.append(js['owner']['owner_uid'])
        prd_owner_type.append(js['owner']['owner_type'])
        prd_owner_name.append(js['owner']['name'])
    except:
        continue
```

```python
dmd_df = pd.DataFrame({
        'dmd_id': dmd_id
        ,'dmd_title' : dmd_title
        ,'dmd_urlIdentifier' :dmd_urlIdentifier 
        ,'dmd_description':dmd_description
        ,'dmd_key_words' : dmd_key_words
        ,'dmd_expiration_date' : dmd_expiration_date
})
```

```python
prd_df = pd.DataFrame({
        'prd_id': prd_id[:17001]
        ,'prd_title' : prd_title[:17001]
        ,'prd_urlIdentifier' :prd_urlIdentifier[:17001]
        ,'prd_description':prd_description[:17001]
        ,'prd_key_words' : prd_key_words[:17001]
        ,'prd_owner_id' :prd_owner_id[:17001]
        ,'prd_owner_type' :prd_owner_type[:17001]
        ,'prd_owner_name' :prd_owner_name[:17001]
})
```

```python
dmd_df.to_csv('dmd_df_.csv')
prd_df.to_csv('prd_df_.csv')
```
